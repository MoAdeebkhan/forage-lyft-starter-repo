```python
from datetime import date

#Factory method

class CarFactory:
    
    def __init__(self,curr_date,last_service_date,curr_milage,last_service_milage):
        self.curr_date = curr_date
        self.last_service_date = last_service_date
        self.curr_service_milage = curr_milage
        self.last_service_milage = last_service_milage
        

#car 1

class Calliope(CarFactory):
    
    # can change engine-type and battery-type also
    
    def __init__(self,curr_date,last_service_date,curr_milage,last_service_milage,engine='capulet',battery='splinder'):
        
        super().__init__(curr_date,last_service_date,curr_milage,last_service_milage)
        
        self.engine = engine
        self.battery = battery
        
    #Is serviceable
    
    def is_serviceable(self):
        
        service_threshold = self.last_service_date.replace(year=self.last_service_date.year+2)
        
        if (self.curr_service_milage - self.last_service_milage < 30000) and  (service_threshold< date.today()) :
            return False
        else:
            return True
        

#car 2

class Gllisade(CarFactory):
    
    # can change engine-type and battery-type also
    
    def __init__(self,curr_date,last_service_date,curr_milage,last_service_milage,engine='willough',battery='spindler'):
        
        super().__init__(curr_date,last_service_date,curr_milage,last_service_milage)
        
        self.engine = engine
        self.battery = battery
        
    #Is serviceable
        
    def is_serviceable(self):
        
        service_threshold = self.last_service_date.replace(year=self.last_service_date.year+2)
        
        if (self.curr_service_milage - self.last_service_milage < 60000) and  (service_threshold< date.today()) :
            return False
        else:
            return True
        
#car 3
        
class Palindrome(CarFactory):
    
    # can change engine-type and battery-type also
    
    def __init__(self,curr_date,last_service_date,curr_milage,last_service_milage,engine='sternman',battery='spindler'):
        
        super().__init__(curr_date,last_service_date,curr_milage,last_service_milage)
        
        self.engine = engine
        self.battery = battery
        
    def warning_ligth_is_on(self):
        
        light = True
        if(light):
            return True
        else:
            False
        
    #Is serviceable
    
    def is_serviceable(self):
        
        service_threshold = self.last_service_date.replace(year=self.last_service_date.year+2)
        
        if (self.warning_light_is_on) and  (service_threshold< date.today()) :
            return False
        else:
            return True
        

#car 4

class Rorschach(CarFactory):
    
    # can change engine-type and battery-type also
    
    def __init__(self,curr_date,last_service_date,curr_milage,last_service_milage,engine='willough',battery='nubbin'):
        
        super().__init__(curr_date,last_service_date,curr_milage,last_service_milage)
        
        self.engine = engine
        self.battery = battery
        
    #Is serviceable
        
    def is_serviceable(self):
        
        service_threshold = self.last_service_date.replace(year=self.last_service_date.year+4)
        
        if (self.curr_service_milage - self.last_service_milage < 60000) and  (service_threshold< date.today()) :
            return False
        else:
            return True
        
#car 5

class Thovex(CarFactory):
    
    # can change engine-type and battery-type also
    
    def __init__(self,curr_date,last_service_date,curr_milage,last_service_milage,engine='capulet',battery='nubbin'):
        
        super().__init__(curr_date,last_service_date,curr_milage,last_service_milage)
        
        self.engine = engine
        self.battery = battery
        
    #Is serviceable
    
    def is_serviceable(self):
        
        service_threshold = self.last_service_date.replace(year=self.last_service_date.year+4)
        
        if (self.curr_service_milage - self.last_service_milage < 30000) and  (service_threshold< date.today()) :
            return False
        else:
            return True
```
